NumPy 배열만으로 다음 그림을 그려랴.

#### 1.1

다음 그림을 콘솔에 표시하라.

영상의 크기는 가로 501, 세로 701이고, 원의 반지름은 적절한 크기로 결정한다.

<img src='circle_1.png' width="30%"/>

```python
# 필요한 모듈 임포트
import numpy as np
import matplotlib.pyplot as plt
# 원의 컬러 시안 정의
CYAN = [0, 255, 255]
# 영상 크기 정의(701*501)
x = np.full((701, 501, 3), [0,0,0], np.uint8)
# 좌표 250,250을 중심으로 반지름 200짜리 시안색 원을 그림
for r in range(701):
    for c in range(501):
        if (r-250)**2 + (c-250)**2 <= 200**2:
            x[r, c] = CYAN
# 영상에 그린 데이터를 시각화
plt.imshow(x)
plt.show()
```

### 문제 2.

'student_performance.csv' 파일을 읽고 다음 질문에 답하라.


#### 2.1

이 파일에는 모두 76 명의 학생들의 8 번의 과제 성적을 정리한 내용이 담겨있다.
1, 2, 3, 4 학년별 학생 수를 구하여 막대 그래프로 표시하라.

가로 축에는 학년을, 세로 축에는 학생 수를 표시해야 한다.

그래프의 제목, x 축 라벨, y 축 라벨을 모두 표시해야 한다.

```python
import pandas as pd
import matplotlib.pyplot as plt
# CSV 파일 읽어오기
df = pd.read_csv('student_performance.csv')
# 각 학년별 학생 수 계산
student_count_by_grade = df['학년'].value_counts().sort_index()
# 막대 그래프로 표시
plt.bar(student_count_by_grade.index, student_count_by_grade)
plt.xlabel('학년')
plt.ylabel('학생 수')
plt.title('학년별 학생 수')
plt.show()
```

#### 2.2

2학년 학생들의 각 과제별 평균을 구한 후에 막대 그래프로 표시하라.

그래프는 다음과 같아야 한다.

<img src='sophomore_average.png' width="60%"/>

```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽어오기
df = pd.read_csv('student_performance.csv')
# 2학년 데이터 필터링
df_2nd_year = df[df['학년'] == 2]
# 과제 평균 계산
average_scores = df_2nd_year[['과제1', '과제2', '과제3', '과제4', '과제5', '과제6', '과제7', '과제8']].mean()
# 막대 그래프로 표시
plt.rcParams['font.family'] = 'NanumGothic'
plt.bar(average_scores.index, average_scores)
plt.xlabel('과제')
plt.title('2학년 학생 과제별 평균 점수')
plt.show()
```
